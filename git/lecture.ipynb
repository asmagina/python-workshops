## Системы контроля версия и почему среди них популярен Git

### Что такое системы контроля (СКВ) версий и с чем их едят?

CКВ &mdash; это система, которая регистрирует все
изменения в файлах, а в дальнейшем позволяет вернуться к их более ранним
версиям и определить, кем и когда были сделаны конкретные изменения.

### Мотивация в использовании СКВ

* необходимость отката
* просмотр истории эволюции кода из IDE
* защита от потери сорцов
* простой деплой между серверами
* ветвление версий
* руководство разработкой

### Почему Git?

Все плюсы распределенной СКВ:

* GIT сохраняет локально полную копию репозитория. 
* Не нужно иметь доступа к серверу, чтобы сохранить в системе контроля версий измененный файл

Git vs SVN:
* Git хранит «слепки» файлов в сжатом виде.
* Решено большое количество проблем со слиянием веток, которыми богат Subversion

С самого начала Git боролся за популярность с Mercurial (Hg), который в некоторых аспектах проще, но в итоге вырвался в лидеры Git.

Немаловажную роль во всеобщей популярности Git’a в мире CVS сыграл ресурс GitHub, на котором размещены тысячи репозиториев на различных языках.

## Основы ветвления в Git

### Что такое ветвление и зачем оно нужно?

Ветвеление  позволяет изолированно работать над задачами в проекте.

По умолчанию разработка ведется в ветке master.

```python
Image("images/branches.png")
```

### Ветвление, пример использованияи

* Поработаем над проектом.
* Создадим ветку для работы над новой задачей.
* Выполним некоторую работу на этой ветке.
* На этом этапе вам поступит звонок о том, что сейчас критична другая проблема, и её надо срочно решить. Мы сделаем следующее:

* Вернёмся на ветку для версии в производстве.
* Создадим ветку для исправления ошибки.
* После тестирования ветки с исправлением сольём её обратно и отправим в продакшн.
* Вернёмся к своей исходной задаче и продолжим работать над ней.

Без веток практически немыслима совместная разработка -- иначе как добавлять и тестировать новые "фичи" в проект, не ломая при этом основной код?!

### Команды для работы с ветками

Создать новую ветку и перейти в нее:

```python
git chekcout -b mybranch
```

Просмотреть список веток:

```python
git branch
```

Залить изменения в ветку:

```python
git push origin mybranch
```

Перейти в другую ветку (мастер):

```python
git checkout master
```

Слить ветку с мастером:

```python
git merge mybranch
```

Удалить ветку локально:

```python
git branch -D mybranch
```

Удалить ветку ветку напрямую с сервера:

```python
git push origin --delete mybranch
```

Вспомогательное:

```python
git branch -v
git branch --merged
git branch --no-merged
```

### Тэги

Для релизов программного обеспечения принято использовать не ветки, а тэги -- это "снимок" состояния ветки (обычно мастера) в конкретный момент времени.

```python
git tag 1.0.0 1b2e1d63ff
```

В тэг нельзя внести измеения.
